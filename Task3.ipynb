```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('student_feedback.csv')
```

```python
df
```

```python
df_cleaned = df.drop(columns=["S/N", "Student ID"])
```

```python
df_cleaned
```

```python
average_ratings = df_cleaned.mean().sort_values(ascending=False)
```

```python
average_ratings
```

```python
plt.figure(figsize=(10, 7))
average_ratings.plot(kind='bar')
plt.title("Average Ratings per Feedback Category")
plt.ylabel("Average Score")
plt.xticks(rotation=35)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(11, 9))
sns.heatmap(df_cleaned.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Feedback Categories")

plt.xticks(rotation=35)
plt.tight_layout()
plt.show()
```

```python
df["Overall Rating"] = df.drop(columns=["Student ID"]).mean(axis=1)
```

```python
df["Overall Rating"].plot.hist(bins=10, color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title("Distribution of Overall Ratings")
plt.xlabel("Average Rating")
plt.tight_layout()
plt.show()
```
